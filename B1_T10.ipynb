# Missing value handling

```python
data.isna().sum()
```

```python
# Filling missing values in columns using their median
```

```python
for i in ['Dis_Mag_Value','Total_Deaths','Total_Affected','Total_Damages']:
    data[i]=data[i].fillna(data[i].median())
data
```

```python
# Filling missing values in columns which are not numerical with mode
```

```python
for i in ['Disaster_Subtype','Event_Name','Location','Dis_Mag_Scale']:
    data[i]=data[i].fillna(data[i].mode()[0])
data
```

```python
# Checking for null values again
```

```python
data.isna().sum()
```

```python
# Dropping columns which are not required
```

```python
data.drop(['ID_No','Disaster_Subtype','ISO','Region','Location','Dis_Mag_Scale','CPI'],axis=1,inplace=True)
data
```

# Modelling

```python
# Splitting the dataset into training and test data
```

```python
y = data['Total_Deaths']
X = data.drop(['Total_Deaths'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
from sklearn.metrics import accuracy_score, f1_score,recall_score,precision_score, classification_report
```

```python
# To get a consolidated test result:
results = {'Classification Model':[], 'F1 Score':[]}
```

```python
# 1. Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model.fit(X_train,y_train)
y_pred=log_model.predict(X_test)
print('Classification Model: {}'.format(log_model.__class__.__name__))
print('F1 Score: ',f1_score(y_test, y_pred, average='macro').round(4))
print('Classification Report: \n',classification_report(y_test, y_pred))

if log_model.__class__.__name__ not in results['Classification Model']:
    results['Classification Model'].append(log_model.__class__.__name__)
    results['F1 Score'].append(f1_score(y_test, y_pred, average='macro').round(4))
```

```python
# 2. KNN
```

```python
# Finding the optimum k value
from sklearn.neighbors import KNeighborsClassifier
values=[]
neighbors=np.arange(1,10)
for k in neighbors:
    log_model=KNeighborsClassifier(n_neighbors=k, metric='minkowski')
    log_model.fit(X_train, y_train)
    y_pred=log_model.predict(X_test)
    score=accuracy_score(y_test, y_pred)
    values.append(score)
    
# Plotting k vs accuracy_score to get optimum k
plt.plot(neighbors, values, 'o-')
plt.title('accuracy_score vs k')
plt.xticks(np.arange(0.0, 11.0, 1.0))
plt.xlabel('k')
plt.ylabel('accuracy_score')
plt.grid()
plt.show()
```

```python
# performing test with k=9

log_model=KNeighborsClassifier(n_neighbors=9, metric='minkowski')
log_model.fit(X_train, y_train)

y_pred = log_model.predict(X_test)

print('Classification Model: {}'.format(log_model.__class__.__name__))
print('F1 Score: ',f1_score(y_test, y_pred, average='macro').round(4))
print('Classification Report: \n',classification_report(y_test, y_pred))

if log_model.__class__.__name__ not in results['Classification Model']:
    results['Classification Model'].append(log_model.__class__.__name__)
    results['F1 Score'].append(f1_score(y_test, y_pred, average='macro').round(4))
```
